```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

def preprocess(text):
    
    text = text.lower()

    
    text = text.translate(str.maketrans('', '', string.punctuation))

   
    words = word_tokenize(text)

  
    stop_words = set(stopwords.words('russian'))
    words = [word for word in words if word not in stop_words]


    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]

    return words
```

```python
#Использую метода preprocess
with open("Далёкая радуга.txt", "r", encoding="utf-8") as file:
    raduga = file.read()

preprocess_raduga = preprocess(raduga)
print(preprocess_raduga)
```

```python
with open("Обитаемый остров.txt", "r", encoding="utf-8") as file:
    ostrov = file.read()

preprocess_ostrov = preprocess(ostrov)
print(preprocess_ostrov)
```
