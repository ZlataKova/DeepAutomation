```python
import spacy
from spacy import displacy

nlp = spacy.load('ru_core_news_sm')

with open("Стажеры.txt", "r", encoding="utf-8") as file:
    text = file.read()

doc = nlp(text)

for token in doc:
    print(token.text, "\t", token.pos_)
```

```python
displacy.render(doc, style="dep", jupyter=True)
```

```python
from natasha import Segmenter, NewsEmbedding, NewsSyntaxParser, Doc

emb = NewsEmbedding()

segmenter = Segmenter()

syntax_parser = NewsSyntaxParser(emb)

text = 'Подкатил громадный красно-белый автобус.Отлетающих  пригласили садиться.'

doc = Doc(text)

doc.segment(segmenter)

for tocken in doc.tokens:
    print(tocken)
```

```python
doc.parse_syntax(syntax_parser)

for tocken in doc.tokens:
    print(tocken)
```
